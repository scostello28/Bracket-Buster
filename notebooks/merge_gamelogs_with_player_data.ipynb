```python
def ID(row):
    row['ID'] = ",".join([row['Team'], str(row['Season'])])
    return row
```

```python
# Change Tm column to Team
gamelog_cols = gamelogs.columns.tolist()
gamelog_cols = [i if i!= 'Tm' else 'Team' for i in gamelog_cols]
gamelogs.columns = gamelog_cols
```
