```python
def final_p(row):
    if row['Tm'] == row['Team']:
        row['final_p'] = row['p']
    elif row['Tm'] == row['Team_v']:
        row['final_p'] = row['p_v']
    return row
```

```python
df = df.apply(final_p, axis=1)
```
