```python
gl_data = pd.read_pickle('../model_data/gamelog_5_exp_clust.pkl')
odds_data = pd.read_pickle('../data/odds_data.pkl')
```

```python
def select_odds_col(df):
    df = df[['Date', 'Team', 'p', 'Team_v', 'p_v']]
    return df
```

```python
odds = select_odds_col(odds_data)
```

```python
odds.info()
```

```python
def odds_merge_id(row):
    '''
    ID generator to merge odds data with gamelog data for odds dataframe
    '''
    row['ID'] = '{},{}'.format(",".join(sorted([row['Team'], row['Team_v']])), row['Date'])
    return row
```

```python
odds = odds.apply(odds_merge_id, axis=1)
```

```python
def gamelogs_odds_merge_id(row):
    '''
    ID generator to merge odds data with gamelog data for gamelog dataframe
    '''
    # row['matchup'] = ",".join(sorted([row['Team'], row['Team_v']]))
    row['Season'] = row['GameType'][-4:]
#     row['matchup'] = 
    row['ID'] = '{},{}'.format(",".join(sorted([row['Tm'], row['OPTm']])), row['Season'] + row['Date'][-6:])
    return row
```

```python
gl = gl_data.apply(gamelogs_odds_merge_id, axis=1)
```

```python
gl.head()
```

```python
def merge(df1, df2):
    df2.drop(['Date'], axis=1, inplace=True)
    merged_df = pd.merge(df1, df2, how='left', on='ID')
    return merged_df
```

```python
merged_df = merge(gl, odds)
```
