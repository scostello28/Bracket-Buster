## Basic Regression

```python
# Set up features and targets
X = Xy.iloc[:, 1:]
y = Xy.iloc[:, 0]
```

```python
'''Train test split'''
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
'''Standardize Data'''
scale = StandardScaler()
scale.fit(X_train)
X_train = scale.transform(X_train)
X_test = scale.transform(X_test)
```

```python
X_tr = pd.DataFrame(X_train)
features = X_tr.columns.tolist()
axes = X_tr[features].hist(bins=20, normed=1,
                grid=0, edgecolor='none',
                figsize=(20, 20),
                layout=(5,6))

for ax in axes.flatten():
    ax
```

```python
'''Fit model on training data'''
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
'''predict on testing data'''
y_hat = model.predict(X_test)
```

```python
accuracy = metrics.accuracy_score(y_test, y_hat)
precision = metrics.precision_score(y_test, y_hat)
recall = metrics.recall_score(y_test, y_hat)
f1 = metrics.f1_score(y_test, y_hat)
# roc_auc_score = metrics.roc_auc_score(y_test, y_hat)
print('Accuracy: {:.4f} (% predicted correctly)'.format(accuracy))
print('Precision: {:.4f} (predicted positives % correct)'.format(precision))
print('Recall: {:.4f} (% of positives predicted correctly)'.format(recall))
print('f1 Score: {:.4f} (weighted average of Precision and Recall)'.format(f1))
```
